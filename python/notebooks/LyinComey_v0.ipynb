```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import json

from scipy.fftpack import fft, rfft, fftfreq
from scipy.io.wavfile import write
from scipy import log10
from pylab import *

%matplotlib inline

plt.rcParams["figure.figsize"] = (12,9)
```

```python
# specify data directory
data_dir = 'lyincomey'
```

```python
os.chdir(data_dir)
```

```python
files = os.listdir()

df = pd.DataFrame()

# loop through files to read out data
for file in files:
    
    print('Reading in file: ', file)
    
    # read out json data
    with open(file) as jsonfile:
        
        data = json.load(jsonfile)
        
        # pull out only statuses to put into dataframe
        frame = pd.DataFrame.from_records(data['statuses'])
        df = df.append(frame)
        
```
